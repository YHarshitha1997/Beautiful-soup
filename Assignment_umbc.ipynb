```python
from bs4 import BeautifulSoup
import re
infile = open("xml_containing_html.xml","r")
contents = infile.read()
soup = BeautifulSoup(contents,'xml')
files = soup.find_all('content')
i=0
for file in files:
    i+=1
    x = re.findall('href',str(file))
    print("Html file ",i,":",len(x))
    
```
